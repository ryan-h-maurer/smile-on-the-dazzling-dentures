```python
n_unique_CallerNum = smile_outcomes.loc[~smile_outcomes["CallerNum"].isin([-1,-2])]["CallerNum"].nunique()
list_of_unique_Caller_Num = smile_outcomes.loc[~smile_outcomes["CallerNum"].isin([-1,-2])]["CallerNum"].unique()

def make_CallerNum_dict(dataframe, colname):
    results = {}
    x=0
    while x in range (0,n_unique_CallerNum):
        patientdf = dataframe.loc[dataframe[colname]==list_of_unique_Caller_Num[x]]
        results[list_of_unique_Caller_Num[x]] = patientdf
        x += 1
    return results
    
callerNum_dict = make_CallerNum_dict(smile_outcomes,"CallerNum")
```

```python
for key in callerNum_dict:
    working_df = callerNum_dict[key]
    if working_df["Base-line Oral and Conditions Questions - How would you describe the condition of your mouth and teeth?"].nunique() == 1:
        same[key] = working_df
    elif working_df["Base-line Oral and Conditions Questions - How would you describe the condition of your mouth and teeth?"].nunique() > 1:
        change[key] = working_df
    else:
        probblank[key] = working_df
        
# print(same.head())
print(change.keys())
# print(probblank)
```
